```
from darkflow.net.build import TFNet
import cv2

#options = {"model": "cfg/tiny-yolo-voc-png.cfg", "load": "bin/yolo.weights", "threshold": 0.1}
options = {"model": "cfg/tiny-yolo-voc-png.cfg", "load": 15000, "threshold": .05}


tfnet = TFNet(options)


imgcv = cv2.imread("./test/273_59_22062_69_2018-02-16T16-52-30-442Z.jpg")
result = tfnet.return_predict(imgcv)
print(result)

imgcv = cv2.imread("./test/360_59_136761_69_2018-02-20T12-26-50-468Z.jpg")
result = tfnet.return_predict(imgcv)
print(result)

imgcv = cv2.imread("./test/6728_58_242241_62_2018-02-05T05-33-11-360Z.jpg")
result = tfnet.return_predict(imgcv)
print(result)

imgcv = cv2.imread("./test/6729_59_242323_62_2018-02-05T10-28-23-515Z.jpg")
result = tfnet.return_predict(imgcv)
print(result)

imgcv = cv2.imread("./test/6729_59_242527_62_2018-02-05T05-42-20-290Z.jpg")
result = tfnet.return_predict(imgcv)
print(result)


imgcv = cv2.imread("./test/pharma_5.jpg")
result = tfnet.return_predict(imgcv)
print(result)
```
